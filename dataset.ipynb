```python
import kagglehub
import os
import shutil
import pandas as pd
import pandas as pd
from IPython.display import display, HTML


# Download dataset
path = kagglehub.dataset_download("ibamibrahim/indonesian-news-title")

# Pastikan folder tujuan ada
os.makedirs("dataset", exist_ok=True)

# Pindahkan hasil download ke folder 'dataset'
for item in os.listdir(path):
    src = os.path.join(path, item)
    dst = os.path.join("dataset", item)
    if os.path.isdir(src):
        shutil.copytree(src, dst, dirs_exist_ok=True)
    else:
        shutil.copy2(src, dst)

print("Dataset berhasil dipindahkan ke folder 'dataset/'")
```

```python
data_path = os.path.join("dataset", "indonesian-news-title.csv")
df = pd.read_csv(data_path)
df.head()
```

```python
import csv
with open("dataset/indonesian-news-title.csv", encoding="utf-8") as f:
    r = csv.reader(f)
    for _ in range(5):
        print(next(r))
```

```python
df['category'].value_counts()
```

```python
from sklearn.utils import resample

# group by label
groups = [df[df['category'] == label] for label in df['category'].unique()]

# cari jumlah kelas terkecil
min_size = min(len(g) for g in groups)
print("Ukuran target:", min_size)

df_balanced = pd.concat([
    g.sample(min_size, random_state=42)
    for g in groups
], ignore_index=True)

df_balanced['category'].value_counts()
```

```python
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
df_balanced.head()

```

```python
os.makedirs("dataset", exist_ok=True)
df_balanced.to_csv(os.path.join("dataset", "indonesian-news-title-balanced.csv"), index=False)
```
